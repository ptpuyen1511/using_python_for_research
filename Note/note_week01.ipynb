```python
def distance(a, b):
    return m.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)

distance((0,0), (1,1))
```

```python
def in_circle(x, origin = [0,0]):
    if distance(x, origin) < 1:
        return True
    return False
```

```python
in_circle((1, 1))
```

```python
R = 10000
inside = []
for i in range(R):
    point = (rand(), rand())
    inside.append(in_circle(point))

print(sum(inside)/R)
```

```python
m.pi/4 - sum(inside)/R
```

```python
import random

random.seed(1)

def moving_window_average(x, n_neighbors=1):
    n = len(x)
    width = n_neighbors*2 + 1
    x = [x[0]]*n_neighbors + x + [x[-1]]*n_neighbors
    return [sum(x[i:(i+width)]) / width for i in range(n)]
```

```python
x=[0,10,5,3,1,5]

sum(moving_window_average(x))
```

```python
moving_window_average(x)
```

```python
random.seed(1)

R = 1000
x = [random.random() for i in range(R)]
Y = [x] + [moving_window_average(x, i) for i in range(1, 10)]

Y[5][9]
```
